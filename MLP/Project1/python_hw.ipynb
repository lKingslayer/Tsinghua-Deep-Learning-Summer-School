```python
import numpy as np
import torch

array = np.arange(20)
array1 = np.copy(array)
print("input:",array)
print("odd numers:",array[1:20:2])
        
for index in range(20):
    if (array[index] % 2 == 1):
        array[index] = -1
print("modify odd numbers:",array)

for index in range(20):
    if (array1[index] % 2 == 0):
        array1[index] = -1
print("modify even numbers:",array1)
```

Problem 2:

```python
matrix = np.arange(1,17).reshape(4,4)
print("Original matrix:")
print(matrix)

print("Take out the middle 2×2 matrix:")
print(matrix[1:3,1:3])

print("Switch the first and second columns of the matrix:")
temp = np.copy(matrix)
temp[:,[0,1]] = temp[:,[1,0]]
print(temp)

print("Reverse each row in the matrix:")
reversedarr = np.flip(matrix, axis=1)
print(reversedarr)
```
